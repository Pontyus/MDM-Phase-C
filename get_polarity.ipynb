```python
import pandas as pd
from textblob import TextBlob
```

```python
tweets1 = pd.read_csv('tweets1.csv')
tweets2 = pd.read_csv('tweets2.csv')
df = tweets1.append(tweets2)
```

```python
df['Datetime'] = df['Datetime'].astype(str)
df['Datetime'] = df['Datetime'].apply(lambda  x:x[0:13])

df.head()
```

```python
df = df[~df['Text'].isnull()]
df.dropna(axis = 0, how = 'any', inplace = True) 

tweets = df['Text'].values
```

```python
polarity = []
subjectivity = []
i = 0
for tweet in tweets:
    if tweet:
        blob = TextBlob(tweet)
        polarity.append(blob.sentences[0].polarity)
        subjectivity.append(blob.sentences[0].subjectivity)
```

```python
df["polarity"] = polarity
df["subjectivity"] = subjectivity
df.head()
```

```python
df.to_csv("datasets_with_polarity.csv")
```
